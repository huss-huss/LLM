```python
import os
import PyPDF2

def extract_text_from_pdf(pdf_path):
    try:
        with open(pdf_path, 'rb') as file:
            reader = PyPDF2.PdfReader(file)
            text = ""
            for page in reader.pages:
                text += page.extract_text()
        return text
    except Exception as e:
        print(f"Error reading {pdf_path}: {e}")
        return None

def process_pdfs(pdf_folder, output_folder):
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    for filename in os.listdir(pdf_folder):
        if filename.endswith('.pdf'):
            pdf_path = os.path.join(pdf_folder, filename)
            print(f"Processing file: {pdf_path}")  # Debugging print
            text = extract_text_from_pdf(pdf_path)
            if text is not None:
                output_filename = os.path.splitext(filename)[0] + '.txt'
                output_path = os.path.join(output_folder, output_filename)
                with open(output_path, 'w', encoding='utf-8') as f:
                    f.write(text)
                print(f"Generated text file: {output_path}")  # Debugging print
            else:
                print(f"Skipping file due to error: {pdf_path}")


# English

# pdf_folder = r'Education\English\Texts'  
# output_folder = r'Txt\English\Texts'

# pdf_folder = r'Education\English\Worksheets'  
# output_folder = r'Txt\English\Worksheets'

# Biology

# pdf_folder = r'Education\Biology\Texts'  
# output_folder = r'Txt\Biology\Texts'

# pdf_folder = r'Education\Biology\Worksheets'  
# output_folder = r'Txt\Biology\Worksheets'

# Chemistry

# pdf_folder = r'Education\Chemistry\Texts'  
# output_folder = r'Txt\Chemistry\Texts'

# pdf_folder = r'Education\Chemistry\Worksheets'  
# output_folder = r'Txt\Chemistry\Worksheets'

# Physics

# pdf_folder = r'Education\Physics\Texts'  
# output_folder = r'Txt\Physics\Texts'

# pdf_folder = r'Education\Physics\Worksheets'  
# output_folder = r'Txt\Physics\Worksheets'

# Civics

# pdf_folder = r'Education\Civics\Texts'  
# output_folder = r'Txt\Civics\Texts'

# pdf_folder = r'Education\Civics\Worksheets'  
# output_folder = r'Txt\Civics\Worksheets'

# Maths

# pdf_folder = r'Education\Maths\Texts'  
# output_folder = r'Txt\Maths\Texts'

# pdf_folder = r'Education\Maths\Worksheets'  
# output_folder = r'Txt\Maths\Worksheets'

# Geography

# pdf_folder = r'Education\Geography\Texts'  
# output_folder = r'Txt\Geography\Texts'

# pdf_folder = r'Education\Geography\Worksheets'  
# output_folder = r'Txt\Geography\Worksheets'

# History

# pdf_folder = r'Education\History\Texts'  
# output_folder = r'Txt\History\Texts'

# pdf_folder = r'Education\History\Worksheets'  
# output_folder = r'Txt\History\Worksheets'

# Amharic

# pdf_folder = r'Education\Amharic\Books'  
# output_folder = r'Txt\Amharic\Texts'


process_pdfs(pdf_folder, output_folder)

```

```python
import pytesseract
from pdf2image import convert_from_path
import os

def ocr_from_pdf(pdf_path, lang='amh'):
    try:
        # Specify the full path to the poppler bin directory
        poppler_path = r'poppler-24.02.0\Library\bin'  # Update this to your actual path
        images = convert_from_path(pdf_path, 500, poppler_path=poppler_path)
        text = ""
        for pageNum, imgBlob in enumerate(images):
            text += pytesseract.image_to_string(imgBlob, lang=lang) + "\n"
        return text
    except Exception as e:
        print(f"Error reading {pdf_path} with OCR: {e}")
        return None

def process_pdfs(pdf_folder, output_folder, lang='amh'):
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    for filename in os.listdir(pdf_folder):
        if filename.endswith('.pdf'):
            pdf_path = os.path.join(pdf_folder, filename)
            print(f"Processing file: {pdf_path}")  # Debugging print
            text = ocr_from_pdf(pdf_path, lang)
            if text is not None:
                output_filename = os.path.splitext(filename)[0] + '.txt'
                output_path = os.path.join(output_folder, output_filename)
                with open(output_path, 'w', encoding='utf-8') as f:
                    f.write(text)
                print(f"Generated text file: {output_path}")  # Debugging print
            else:
                print(f"Skipping file due to error: {pdf_path}")

# Use raw string (r'') to handle backslashes in Windows paths

# Example usage for Amharic PDFs
pdf_folder = r'C:\Users\husib\OneDrive\Desktop\Data Preprocessing\Education\Amharic\Texts'
output_folder = r'C:\Users\husib\OneDrive\Desktop\Data Preprocessing\Txt\Amharic\Texts'

process_pdfs(pdf_folder, output_folder, lang='amh')
```
